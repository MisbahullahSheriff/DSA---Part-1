#### 2D Operations

##### creation

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(type(matrix))
```

##### display

```python
for row in matrix:
    print(row)
```

```python
def display_matrix(mat):
    for row in mat:
        print(row)
```

```python
display_matrix(matrix)
```

##### traversal

```python
for row in matrix:
    for ele in row:
        print(ele, end=" ")
    print()
```

```python
m = len(matrix)
n = len(matrix[0])

for i in range(m):
    for j in range(n):
        print(matrix[i][j], end=" ")
    print()
```

```python
print(matrix[2][1])
```

##### updation

```python
matrix[1][2] = 10
display_matrix(matrix)
```

### Arrays

#### 1. Array

```python
import array
```

##### creation

```python
arr1 = array.array('i', [10, 20, 30, 40, 50])
print(arr1)
```

```python
arr2 = array.array('i', [60, 70, 80])
print(arr2)
```

##### access

```python
arr1[2]
```

```python
arr1[3]
```

##### traversal

```python
for element in arr1:
    print(element, end=" ")
```

##### insertion

```python
arr1.insert(3, 35)
```

```python
print(arr1)
```

##### deletion

```python
arr1.remove(30)
print(arr1)
```

##### updation

```python
arr1[2] = 30
print(arr1)
```

##### sorting

```python
arr3 = array.array('i', [4, 7, 3, 9, 2, 6])
print(arr3)
```

```python
sorted_arr3 = array.array('i', sorted(arr3))
print(sorted_arr3)
```

##### merging

```python
arr1
```

```python
arr2
```

```python
arr4 = arr1 + arr2
arr4
```

##### splitting

```python
ix = int(len(arr4) / 2)
ix
```

```python
arr4[:ix]
```

```python
arr4[ix:]
```

#### 2. Numpy

```python
import numpy as np
```

##### creation

```python
np1 = np.array([10, 20, 30, 40, 50])
np2 = np.array([50, 60, 70])

print(np1)
print(np2)
```

```python
np1
```

##### access

```python
np1[0]
```

##### traversal

```python
for ele in np1:
    print(ele, end=" ")
```

##### insertion

```python
np.insert(np1, 3, 35)
```

##### deletion

```python
np.delete(np1, 3)
```

##### updation

```python
np1[2] = 35
print(np1)
```

##### sorting

```python
np3 = np.array([5, 3, 7, 6, 2, 9])
np3
```

```python
np.sort(np3)
```

##### merging

```python
np1
```

```python
np2
```

```python
np3 = np.concatenate((np1, np2))
np3
```

##### splitting

```python
split1, split2 = np.array_split(np3, 2)
```

```python
split1
```

```python
split2
```

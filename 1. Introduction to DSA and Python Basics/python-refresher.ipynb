### 6. List Comprehensions

```python
result = []

for i in range(1, 11):
    if i % 2 == 0:
        result.append(i)

result
```

```python
list(range(2, 11, 2))
```

```python
seq = range(1, 11)

[a for a in seq if a % 2 == 0]
```

```python
ages
```

```python
for age in ages:
    if age > 30:
        print('Greater than 30')
    else:
        print('Less than 30')
```

```python
['Greater than 30' if age > 30 else 'Less than 30' for age in ages]
```

### 7. Classes

```python
class MyRectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
        self.random_attribute = 0
        self.another_random_attribute = 'abc'

    def get_perimeter(self):
        return 2 * (self.length + self.breadth)

    def get_area(self):
        return self.length * self.breadth

    def get_length(self):
        return self.length

    def get_breadth(self):
        return self.breadth
```

```python
obj1 = MyRectangle(length=5, breadth=6)
```

```python
obj2 = MyRectangle(length=10, breadth=12)
```

```python
obj1
```

```python
obj1.breadth
```

```python
obj1.random_attribute
```

```python
obj1.another_random_attribute
```

```python
obj1.get_length()
```

```python
MyRectangle.get_length(obj1)
```

```python
MyRectangle.get_length(obj2)
```

```python
obj1.get_area()
```

```python
obj2.get_area()
```

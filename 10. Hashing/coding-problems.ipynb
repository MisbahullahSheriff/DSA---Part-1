## Coding Problems

### 1. Character Count

```python
def char_count(text):
    freq = dict()
    for ch in text:
        freq[ch] = freq.get(ch, 0) + 1
    return freq
```

```python
char_count('this is a text')
```

```python
char_count('Mississippi')
```

### 2. First Non-repeating Character

```python
def first_non_repeating_char(text):
    freq = dict()
    for ch in text:
        freq[ch] = freq.get(ch, 0) + 1
    for k, v in freq.items():
        if v == 1:
            return k
    return None
```

```python
first_non_repeating_char('Mississippi')
```

```python
first_non_repeating_char('this is a text')
```

### 3. Two-Sum Problem

```python
def two_sum(lst, target):
    seen = set()
    for x in lst:
        if (target - x) in seen:
            return (x, target - x)
        seen.add(x)
    return None
```

```python
two_sum([3, 4, 4, 3, 2, 1, 5, 6, 2], target=10)
```

```python
two_sum([3, 4, 4, 3, 2, 1, 5, 6, 2], target=5)
```

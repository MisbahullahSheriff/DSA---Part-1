## Hashing Functions

### 1. Simple Hash

```python
def simple_hash(key):
    total = 0
    for ch in key:
        total += ord(ch)
    return total
```

```python
simple_hash('ball'), simple_hash('bat')
```

### 2. Multiplicative Hash

```python
def multiplicative_hash(key):
    h = 1
    for ch in key:
        h = (h * 31) + ord(ch)
    return h
```

```python
multiplicative_hash('ball'), multiplicative_hash('bat')
```

### 3. DJB2

```python
def djb2(s):
    h = 5381
    for ch in s:
        (h << 31) + ord(ch)
    return h
```

```python
djb2('ball'), djb2('apple')
```

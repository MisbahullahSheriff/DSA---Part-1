## Linear Search

```python
def linear_search(arr, target):
    for index, element in enumerate(arr):
        if element == target:
            return index
    return -1
```

```python
data = [4, 6, 1, 3, -1, 10, 5, 8]
```

```python
linear_search(data, -1)
```

```python
linear_search(data, 5)
```

```python
linear_search(data, 100)
```

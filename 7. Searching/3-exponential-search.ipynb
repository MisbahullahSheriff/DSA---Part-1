## Exponential Search

```python
def binary_search(arr, target, low, high):
    while low <= high:
        mid = low + (high - low) // 2
        if arr[mid] == target:
            return mid
        if arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1


def exponential_search(arr, target):   
    if arr[0] == target:
        return 0
    
    i = 1
    n = len(arr)
    while (i < n) and (arr[i] <= target):
        i *= 2
    
    low = i // 2
    high = min(i, n - 1)
    print(f'Target: {target}')
    print(f'Range: ({low}, {high})')
    return binary_search(arr, target, low, high)
```

```python
data = [10, 20, 30, 40, 50, 60, 70, 100, 120, 150, 180, 200, 250, 300, 500]
```

```python
exponential_search(data, 250)
```

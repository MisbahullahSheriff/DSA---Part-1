## Bubble Sort

```python
def bubble_sort(arr):
    n =  len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
```

```python
seq = [10, 5, 8, 3, 2, 1, 9]
```

```python
bubble_sort(seq)
```

## Count Sort

```python
def count_sort(arr):
    max_val = max(arr)
    count_arr = [0] * (max_val + 1)

    for ele in arr:
        count_arr[ele] += 1
    
    result = []
    for i in range(len(count_arr)):
        result.extend([i] * count_arr[i])
    
    return result
```

```python
data = [5, 3, 8, 2, 3, 4, 8, 4, 5, 5, 2, 3]
```

```python
count_sort(data)
```

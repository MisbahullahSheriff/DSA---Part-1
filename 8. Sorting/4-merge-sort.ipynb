## Merge Sort

```python
def merge(left, right):
    result = []
    i = j = 0

    while (i < len(left)) and (j < len(right)):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    result.extend(left[i:])
    result.extend(right[j:])

    return result
```

```python
def merge_sort(arr):
    if len(arr) == 1:
        return arr
    
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)
```

```python
data = [20, 10, 15, 7, 5]

merge_sort(data)
```

```python
data = [20, 10, 15, 7, 5, 30, 40, 25, 17, 9, 2]

merge_sort(data)
```

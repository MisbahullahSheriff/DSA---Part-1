## Selection Sort

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

```python
data = [4, 2, 7, 1, 5, 9, 10]
```

```python
selection_sort(data)
```

```python
data = [4, 2, 7, 1, 5, 9, 10, 3, -4]
selection_sort(data)
```

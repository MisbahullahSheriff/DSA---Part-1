## Greedy Approach

### 1. Activity Selection (Interval Scheduling)

- Select maximum number of non-overlapping activities

```python
def activity_selection(intervals):
    intervals.sort(key=lambda val: val[1])

    count = 0
    last_end = -float('inf')

    for start, end in intervals:
        if start >= last_end:
            count += 1
            last_end = end
    
    return count
```

```python
intervals = [(1,3), (2,4), (3,5), (0,6), (5,7)]

activity_selection(intervals)
```

### 2. Minimum Platforms (Railway station problem)

- Given arrival & departure times of trains, find minimum platforms needed so no train waits

```python
def min_platforms(arrivals, departures):
    arrivals.sort()
    departures.sort()

    i = j = 0
    curr_platforms = 0
    result = 0

    while i < len(arrivals):
        if arrivals[i] <= departures[j]:
            curr_platforms += 1
            result = max(result, curr_platforms)
            i += 1
        else:
            curr_platforms -= 1
            j += 1
    
    return result
```

```python
arrivals = [900, 940, 950, 1100]
departures = [910, 1200, 1120, 1130]

min_platforms(arrivals, departures)
```

### 3. Jump Game

- Given an array `arr`, each element represents max jump length
- Return whether you can reach last index

```python
def can_jump(arr):
    farthest = 0

    for i in range(len(arr)):
        if i > farthest:
            return False
        farthest = max(farthest, i + arr[i])
    
    return True
```

```python
print(can_jump([2,3,1,1,4]))
print(can_jump([3,2,1,0,4]))
```

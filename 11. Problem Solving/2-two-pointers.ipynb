## Two Pointers Technique

### 1. Remove Duplicates from Sorted Array

- Given a sorted array, remove duplicates in place and return the length of the new array containing unique elements

- `Input`: [1,1,2,2,3]
- `Output`: 3

```python
def remove_duplicates(arr):
    i = 0
    for j in range(1, len(arr)):
        if arr[i] != arr[j]:
            i += 1
            arr[i] = arr[j]
    return i + 1, arr[: i+1]
```

```python
remove_duplicates([1, 1, 2, 2, 3])
```

### 2. Find All Unique Triplets

- Given array `arr`, return all unique triplets (a,b,c) such that `a + b + c = 0`

```python
def find_triplets(arr):
    arr.sort()
    res = []

    for i in range(len(arr) - 2):
        if (i > 0) and (arr[i] == arr[i-1]):
            continue

        left, right = i + 1, len(arr) - 1
        while left < right:
            s = arr[i] + arr[left] + arr[right]
            if s == 0:
                res.append((arr[i], arr[left], arr[right]))
                left += 1
                right -= 1
                while (left < right) and (arr[left] == arr[left - 1]):
                    left += 1
            elif s < 0:
                left += 1
            else:
                right -= 1
    
    return res
```

```python
find_triplets([-1, 1, 0, 2, -1, -4])
```

### 3. Two Sum II (Input Sorted Array)

- Given a sorted array, find two numbers such that they add up to a target

```python
def two_sum(arr, target):
    left, right = 0, len(arr) - 1
    while left < right:
        s = arr[left] + arr[right]
        if s == target:
            return (left, right)
        elif s < target:
            left += 1
        else:
            right -= 1
    return []
```

```python
two_sum([2, 7, 11, 15], 18)
```

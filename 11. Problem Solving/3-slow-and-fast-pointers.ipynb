## Fast & Slow Pointer

### 1. Detect Cycle in Linked List

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution:
    def has_cycle(self, head):
        slow, fast = head, head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False


sol = Solution()
```

```python
# Test Case 1 — Linked List with a Cycle

# create nodes
n1 = ListNode(1)
n2 = ListNode(2)
n3 = ListNode(3)
n4 = ListNode(4)

# Link nodes
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n2

# Test
sol.has_cycle(n1)
```

```python
# Test Case 2 — Linked List without a Cycle

# Create nodes
a1 = ListNode(1)
a2 = ListNode(2)
a3 = ListNode(3)
a4 = ListNode(4)

# Link nodes
a1.next = a2
a2.next = a3
a3.next = a4

# Test
sol.has_cycle(a1)
```

```python
# Test Case 3 — Single Node

single = ListNode(1)

# Test
sol.has_cycle(single)
```

```python
# Test Case 4 — Single Node self-cycle

self_cycle = ListNode(1)
self_cycle.next = self_cycle

# Test
sol.has_cycle(self_cycle)
```

### 2. Find the Start of the Cycle in Linked List

```python
class Solution:
    def __init__(self):
        self.slow = None
        self.fast = None

    def has_cycle(self, head):
        self.slow, self.fast = head, head

        while self.fast and self.fast.next:
            self.slow = self.slow.next
            self.fast = self.fast.next.next
            if self.slow == self.fast:
                return True
            
        return False
    
    def start_cycle(self, head):
        if not self.has_cycle(head):
            return None
        
        self.slow = head
        while self.slow != self.fast:
            self.slow = self.slow.next
            self.fast = self.fast.next
            
        return self.slow
    

sol = Solution()
```

```python
# Test Case 1

a1 = ListNode(1)
a2 = ListNode(2)
a3 = ListNode(3)
a4 = ListNode(4)
a5 = ListNode(5)

a1.next = a2
a2.next = a3
a3.next = a4
a4.next = a5
a5.next = a3

# Test
res = sol.start_cycle(a1)
print(res.val if res else None)
```

```python
# Test Case 2

b1 = ListNode(1)
b2 = ListNode(2)
b3 = ListNode(3)
b4 = ListNode(4)

b1.next = b2
b2.next = b3
b3.next = b4

# Test
res = sol.start_cycle(b1)
print(res.val if res else None)
```

### 3. Find Duplicate Number (`Floyd Cycle Detection on Arrays`)

```python
def find_duplicate(arr):
    slow = arr[0]
    fast = arr[arr[0]]

    while slow != fast:
        slow = arr[slow]
        fast = arr[arr[fast]]

    slow = 0
    while slow != fast:
        slow = arr[slow]
        fast = arr[fast]
    
    return slow
```

```python
find_duplicate([2, 1, 3, 3, 4])
```

## Binary Search Pattern

### 1. First and Last Occurrence in Sorted Array

- Given a sorted array `arr` and a `target`, find the first and last index

```python
def search_occurance(arr, target):
    def first_index():
        left, right = 0, len(arr) - 1
        index = -1

        while left <= right:
            mid = (right + left) // 2
            if arr[mid] >= target:
                right = mid - 1
            else:
                left = mid + 1
            if arr[mid] == target:
                index = mid
        return index

    def last_index():
        left, right = 0, len(arr) - 1
        index = -1

        while left <= right:
            mid = (right + left) // 2
            if arr[mid] <= target:
                left = mid + 1
            else:
                right = mid - 1
            if arr[mid] == target:
                index = mid
        return index
    
    return [first_index(), last_index()]
```

```python
search_occurance([2, 4, 4, 4, 5, 7], 4)
```

```python
search_occurance([-3, -3, -2, -2, 0, 4, 4, 5, 6, 7], target=-2)
```

### 2. Minimum Days to Make Bouquets

- `bloomDay[i]` = day i-th flower blooms
- Need to make `m` bouquets
- Each bouquet needs `k` adjacent flowers

```python
def min_days(bloom_days, m, k):
    if m * k > len(bloom_days):
        return -1
    
    
    def can_make(day):
        count = 0
        bouquets = 0

        for d in bloom_days:
            if d <= day:
                count += 1
            else:
                count = 0
            
            if count == k:
                bouquets += 1
                count = 0
        
        return bouquets >= m


    low, high = min(bloom_days), max(bloom_days)
    result = -1

    while low <= high:
        mid = (low + high)  // 2
        if can_make(mid):
            result = mid
            high = mid - 1
        else:
            low = mid + 1
    
    return result
```

```python
bloom_days = [1,10,3,10,2]
m = 3
k = 1

min_days(bloom_days, m, k)
```

### 3. Aggressive Cows

- Given positions of `stalls` and `number of cows`
- Place cows in stalls such that minimum distance between any 2 cows is maximized

```python
def aggressive_cows(stalls, cows):
    stalls.sort()


    def can_place(distance):
        placed = 1
        last = stalls[0]

        for pos in stalls[1:]:
            if pos - last >= distance:
                placed += 1
                last = pos
            
            if placed == cows:
                return True
        
        return False


    low, high = 1, stalls[-1] - stalls[0]
    result = 0

    while low <= high:
        mid = (low + high) // 2

        if can_place(mid):
            result = mid
            low = mid + 1
        else:
            high = mid - 1
    
    return result
```

```python
stalls = [1,2,8,4,9]
cows = 3

aggressive_cows(stalls, cows)
```

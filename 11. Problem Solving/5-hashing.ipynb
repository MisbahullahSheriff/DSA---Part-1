## Hashing

### 1. Top K Frequent Elements

```python
from collections import Counter
```

```python
def top_frequent(arr, k):
    freq = Counter(arr)
    return sorted(freq, key=lambda x: freq[x], reverse=True)[:k]
```

```python
arr = [3, 1, 2, 2, 2, 5, 5, 5, 1, 5, 3]
top_frequent(arr, 3)
```

### 2. Valid Anagram

```python
def is_anagram(s1, s2):
    if len(s1) != len(s2):
        return False
    return Counter(s1) == Counter(s2)
```

```python
inputs = [
    ('thing', 'night'),
    ('loving', 'caring'),
    ('listen', 'silent'),
    ('inch', 'chin'),
    ('hand', 'band')
]

for inp in inputs:
    print(f'{inp[0]}, {inp[1]}: {is_anagram(inp[0], inp[1])}\n')
```

### 3. Isomorphic Strings

```python
def is_isomorphic(str1, str2):
    if len(str1) != len(str2):
        return False
    
    map1 = dict()
    map2 = dict()

    for c1, c2 in zip(str1, str2):
        if (c1 in map1) and (map1[c1] != c2):
            return False
        if (c2 in map2) and (map2[c2] != c1):
            return False
        map1[c1] = c2
        map2[c2] = c1
        
    return True
```

```python
inputs = [
    ('egg', 'add'),
    ('foo', 'bar'),
    ('paper', 'title'),
    ('ab', 'aa')
]

for inp in inputs:
    print(f'{inp[0]}, {inp[1]}: {is_isomorphic(inp[0], inp[1])}\n')
```

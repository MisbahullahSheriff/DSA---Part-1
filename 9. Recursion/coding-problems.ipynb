## Coding Problems on Recursion

### 1. Print numbers from `N to 1`

```python
def print_numbers(n):
    # base condition
    if n == 0:
        return
    
    print(n, end=' ')
    print_numbers(n - 1)
```

```python
print_numbers(10)
```

### 2. Sum of first `N` natural numbers

```python
def sum_first_natural_numbers(n):
    # base condition
    if n == 0:
        return 0
        
    return n + sum_first_natural_numbers(n - 1)
```

```python
sum_first_natural_numbers(5)
```

```python
sum_first_natural_numbers(4)
```

```python
sum_first_natural_numbers(10)
```

### 3. Count digits of a number

```python
def count_digits(n):
    # base condition
    if n == 0:
        return 0
    
    return 1 + count_digits(n // 10)
```

```python
count_digits(99)
```

### 4. Power function

```python
def get_power(n, k):
    # base condition
    if k == 0:
        return 1
    
    return n * get_power(n, k - 1)
```

```python
get_power(3, 4)
```

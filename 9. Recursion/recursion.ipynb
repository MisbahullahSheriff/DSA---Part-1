## Recursion

### 1. Factorial

```python
def factorial(n):
    if (n == 0) or (n == 1):
        return 1
    return n * factorial(n - 1)
```

```python
factorial(5)
```

### 2. Fibonacci

```python
def fibonacci(n):
    if (n == 0) or (n == 1):
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
fibonacci(6)
```

```python
for i in range(10):
    print(fibonacci(i), end=', ')
```

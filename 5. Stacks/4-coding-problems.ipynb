## Coding Problems using Stacks

```python
from collections import deque
```

### 1. Reverse a String using Stack

```python
def reverse_string_using_stack(string):
    stack = deque()
    result = ''

    # push characters into the stack
    for ch in string:
        stack.append(ch)
    
    # pop each element from the stack
    while stack:
        result += stack.pop()
    
    return result
```

```python
reverse_string_using_stack('This is a string!!')
```

### 2. Check Balanced Parentheses

```python
def is_balanced_parantheses(expression) -> bool:
    stack = deque()
    pairs = {
        ')': '(',
        ']': '[',
        '}': '{'
    }

    for ch in expression:
        if ch in '({[':
            stack.append(ch)
        elif ch in ')}]':
            pair = pairs[ch]
            if (not stack) or pair != stack.pop():
                return False
    
    return len(stack) == 0
```

```python
is_balanced_parantheses('(({[{[]()[]{}}]}))')
```

### 3. Undo Operation Simulation

```python
def simulate_undo_operation(commands):
    stack = deque()

    for command in commands:
        if command == 'undo':
            if stack:
                stack.pop()
            else:
                print('Empty stack.. cannot perform UNDO')
                return
        else:
            stack.append(command[-1])
    
    return ''.join(stack)
```

```python
commands = [
    'type A',
    'type B',
    'undo',
    'type C',
    'type C',
    'type F'
]

res = simulate_undo_operation(commands)
print(f'Result: {res}')
```

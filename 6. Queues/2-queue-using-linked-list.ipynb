### Implementing Queue using Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class QueueUsingLinkedList:
    def __init__(self):
        self.rear = self.front = None

    
    def is_empty(self):
        return self.front is None


    def display(self):
        if self.is_empty():
            print('Cannot display... Queue is empty')
            return
        curr = self.front
        while curr:
            print(curr.data, end=' ')
            curr = curr.next
        print()
    

    def enqueue(self, data):
        node = Node(data)
        if self.is_empty():
            self.rear = self.front = node
        else:
            self.rear.next = node
            self.rear = node
        self.display()

    
    def dequeue(self):
        if self.is_empty():
            print('Cannot dequeue... Queue is empty')
            return
        curr = self.front
        self.front = curr.next
        if not self.front:
            self.rear = None
        dequeued = curr.data
        self.display()
        return dequeued
    

    def front_peek(self):
        if self.is_empty():
            print('Cannot front peek... Queue is empty')
            return
        return self.front.data



    def rear_peek(self):
        if self.is_empty():
            print('Cannot rear peek... Queue is empty')
            return
        return self.rear.data
```

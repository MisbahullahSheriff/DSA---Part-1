### 1. Queue using Stack

```python
class QueueUsingStack:
    def __init__(self):
        self.s1 = deque()
        self.s2 = deque()
    

    def enqueue(self, data):
        self.s1.append(data)
    

    def dequeue(self):
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        if not self.s2:
            print('Queue is empty.. cannot Dequeue')
            return
        return self.s2.pop()
```

### 2. Reverse first `k` in a Queue

- `q`: given queue

```python
# input1: k = 3
q1 = deque()
for x in range(1, 6):
    q1.append(x)

# input2: k = 4
q2 = deque()
for x in range(4, 0, -1):
    q2.append(x)

print(q1, q2)
```

```python
def reverse_first_k(q, k):
    stack = deque()
    result = deque()

    for _ in range(k):
        stack.append(q.popleft())
    
    while stack:
        result.append(stack.pop())

    while q:
        result.append(q.popleft())
    
    return result
```

```python
reverse_first_k(q1, 3)
```

```python
reverse_first_k(q1, 2)
```

```python
reverse_first_k(q2, 4)
```

### 3. Ticket Counter

- `n` people from 1 to `n` are standing in the queue at a movie ticket counter
- Here, tickets are given to the first `k` people and then the last `k` people and again first `k` people and so on
- Once a person gets a ticket moves out of the queue
- The task is to find the last person to get the ticket



```python
def get_last_person(n, k):
    if n == 1:
        return n

    # populating the ticket counter
    q = deque()
    for x in range(1, n + 1):
        q.append(x)

    while True:
        # remove the first k
        for _ in range(k):
            q.popleft()
            if len(q) == 1:
                return q[0]
        
        # remove the last k
        for _ in range(k):
            q.pop()
            if len(q) == 1:
                return q[0]
```

```python
get_last_person(9, 3)
```

```python
get_last_person(5, 1)
```

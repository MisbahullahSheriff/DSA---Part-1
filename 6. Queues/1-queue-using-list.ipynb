### Implementing Queue using List

```python
class QueueUsingList:
    def __init__(self):
        self.queue = []


    def is_empty(self):
        return len(self.queue) == 0
    

    def display(self):
        if self.is_empty():
            print('Cannot display.. Queue is empty')
            return
        print(self.queue)
    

    def enqueue(self, data):
        self.queue.append(data)
        self.display()


    def dequeue(self):
        if self.is_empty():
            print('Queue is empty')
            return 
        
        dequeued = self.queue.pop(0)
        self.display()
        return dequeued


    def front_peek(self):
        if self.is_empty():
            print('Queue is empty')
            return 
        return self.queue[0]


    def rear_peek(self):
        if self.is_empty():
            print('Queue is empty')
            return 
        return self.queue[-1]
```

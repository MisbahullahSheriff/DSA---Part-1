### Implementing Queue using `collections`

```python
from collections import deque
```

```python
class QueueUsingCollections:
    def __init__(self):
        self.queue = deque()

    
    def is_empty(self):
        return len(self.queue) == 0
    

    def display(self):
        if self.is_empty():
            print('Cannot display.. Queue is empty')
        else:
            print(self.queue)
    

    def enqueue(self, data):
        self.queue.append(data)
        self.display()


    def dequeue(self):
        if self.is_empty():
            print('Cannot dequeue.. Queue is empty')
            return
        
        dequeued = self.queue.popleft()
        self.display()
        return dequeued


    def front_peek(self):
        if self.is_empty():
            print('Cannot front peek.. Queue is empty')
            return
        return self.queue[0]



    def rear_peek(self):
        if self.is_empty():
            print('Cannot rear peek.. Queue is empty')
            return
        return self.queue[-1]
```

```python
my_queue = QueueUsingCollections()
my_queue
```
